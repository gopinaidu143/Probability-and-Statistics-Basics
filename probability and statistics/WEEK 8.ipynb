# Descriptive Statistics 
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.
#Attribute (Column) Information:
1. Number of times pregnant
2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
3. Diastolic blood pressure (mm Hg) — Blood Pressure
4. Triceps skinfold thickness (mm)
5. 2-Hour serum insulin (mu U/ml)
6. Body mass index (weight in kg/(height in m)²)
7. Diabetes pedigree function — a function which scores the likelihood of diabetes based on family history
8. Age (years)
9. Class variable (0 or 1) — 0 signifies non-diabetic and 1 signifies diabetic

```python
'''A) Load the dataset and assign the column names
['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class'] 
also display the top 5 rows of the dataset '''

from pandas import read_csv
filename="pima-indians-diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
pima_df = read_csv(filename, names=names)
types = pima_df.dtypes
print(pima_df.shape)
print(pima_df.size)
print(pima_df.head(10))
print(types)
```

```python
# B) Calculating Central Tendencies (Mean,median and Mode)

#Central tendency calculations
print('MEAN: \n',pima_df.mean())
print('\r \n')

print('MEDIAN: \n',pima_df.median())
print('\r \n')

print('MODE: \n',pima_df.mode())
print('\r \n')

print('Q1-Quartile 1: \n',pima_df.quantile(q=(0.25,0.50,0.75)))
```

```python
# C) Plot the Distubution Plot for Attribute 'PLAS'

import matplotlib.pyplot as plt
import seaborn as sns

mean=pima_df['plas'].mean()
median=pima_df['plas'].median()
mode=pima_df['plas'].mode()
fig, ax = plt.subplots(figsize=(5,3)); 
sns.distplot(pima_df['plas']);
plt.title('Distribution plot if Plas variable');
plt.axvline(mean,color='green',label='Mean'); 
plt.axvline(median,color='blue',label='Median');
plt.axvline(mode[0],color='red',label='Mode1')
plt.axvline(mode[1],color='red',label='Mode2')
plt.show()
```

```python
#D) Find the Correlation between Atrributes and plot the Correlation using SEABORN library 
# # Pairplot using s
import matplotlib.pyplot as plt
corr = pima_df.corr()
print(corr)
# sns.heatmap(corr, annot = True);
# fig, ax = plt.subplots()
# fig.set_size_inches(10, 10)
plt.title('Heat Map of Correlation \n')
sns.heatmap(corr,annot=True,cmap='YlGnBu', vmin=-1,vmax=1, linewidths=.10, center=0)
plt.show()
```

```python
#E) Check the Outlies for each attaribute using Boxplot
import matplotlib.pyplot as plt 
plt.figure(figsize=(15,10))
pos=1
for i in pima_df.drop(columns={'class'}).columns:
    plt.subplot(3,3,pos)
    sns.boxplot(pima_df[i])
    pos=pos+1

plt.show()

```
