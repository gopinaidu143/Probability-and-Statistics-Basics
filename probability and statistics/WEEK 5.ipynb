```python
'''
Aim: Simulation of random experiment (e.g. Rolling an n-face dice) and 
    plot the empirical probability of each event, alongside the theoretical probability '''

import matplotlib.pyplot as plt
import random
n=int(input("enter number of faces of dice:"))
k=int(input("enter number of trails:"))
c=[]

for i in range(1,n+1):
    c.append(0)  
for i in range(1,n+1):
    r=random.randint(1,n)
    print(r,end=" ")
    c[r-1]=c[r-1]+1

print("\nEMPIRICAL PROBABILITY:")      
for i in range(n):
    print("probability of",i+1," is ",c[i]/n)
x=[i for i in range(1,n+1)]
print("THEORETICAL PROBABILITY:")
s=[]
for i in range(1,n+1):
    y=1/n
    s.append(y)
print(s)

plt.plot(x,c,"b",label="empirical")
plt.plot(x,s,"r",label="theoretical")
plt.xlabel("outcomes")
plt.ylabel("probability")
plt.title("EP vs TP")
plt.legend(loc="best")
plt.show()
```
