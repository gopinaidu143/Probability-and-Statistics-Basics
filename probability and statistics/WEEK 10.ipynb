Aim: Performing T-test for testingthe Null hypothesis for  difference of meansusing Python
A t-test is a type of inferential statistic which is used to determine if there is a significant difference between 
the means of two groups which may be related in certain features



The average breaking strength of steel rods is specified to be 18.5 thousand pounds. To test this, a sample of 14 rods was tested. The mean and standard deviations obtained were 17.85 and 1.955 respectively. Is the result of experiment significant at 0.05 level of significance

```python




from numpy import sqrt, abs, round
import scipy
from scipy.stats import norm
x_bar=17.85
mu=18.5
s=1.955
n=14
los=0.05
category="two"
df=n-1
def oneSampT(x_bar, mu,s, n, category):
    pooledSE = s/sqrt(n)
    t_calval = (x_bar - mu)/pooledSE
    if category=="two" :
        tab_val = scipy.stats.t.ppf(1-(los/2),df) # two tail # for two tail test
    else:  
        tab_val = scipy.stats.t.ppf(1-los,df)  #  one tail
    return round(t_calval, 3), round(tab_val, 4)

t_calval, tab_val = oneSampT(x_bar, mu, s, n,category)

print (t_calval, tab_val)

if(abs(t_calval) <tab_val):
    print("Accept Null Hypothesis")
else:
    print("Reject NUll Hypothesis")

```

Two-sample T-test With Python:
The Independent Samples t Test or 2-sample t-test compares the means of two independent groups in order to determine 
whether there is statistical evidence that the associated population means are significantly different. 
The Independent Samples t Test is a parametric test. This test is also known as: Independent t Test




An experiment was performed to compare the abrasive wear of two different laminated materials. Twelve pieces of material 1 were tested by exposing each piece to a machine measuring wear. Ten pieces of material 2 were similarly tested. In each case, the depth of wear was observed. The samples of material 1 gave an average (coded) wear of 85 units with a sample standard deviation of 4, while the samples of material 2 gave an average of 81 with a sample standard deviation of 5. Can we conclude at the 0.05 level of significance that the abrasive wear of material 1 exceeds that of material 2 by more than 2 units? Assume the populations to be approximately normal with equal variances.

Null Hypothesis H0: mu1-mu2=2 i.e. mudiff=0 
Alternative Hypothesis H1: mu1-mu2>2 (one tailed test) 
alpha=0.05 


```python
from numpy import sqrt, abs, round
from scipy.stats import norm

X1=85
X2=81
mudiff=2
s1=4
s2=5
n1=12
n2=10
los=0.05
df=n1+n2-2
category="one"
def twoSampZ(X1, X2, mudiff, s1, s2, n1, n2,category):
    S = sqrt(((n1-1)*s1**2 + (n2-1)*s2**2)/(n1+n2-2))
    t_calval = ((X1 - X2) - mudiff)/(S*(sqrt((1/n1)+(1/n2))))
    if category=="two" :
        tab_val = scipy.stats.t.ppf(1-(los/2),df) # two tail # for two tail test
    else:  
        tab_val = scipy.stats.t.ppf(1-los,df)  #  one tail
    return round(t_calval, 3), round(tab_val, 4)

t_calval, tab_val = twoSampZ(X1, X2, mudiff, s1, s2, n1, n2,category)

print (t_calval, tab_val)

if(abs(t_calval) <tab_val):
    print("Accept Null Hypothesis")
else:
    print("Reject NUll Hypothesis")
```

Aim :Write a Python program to perform chi-square test and F-test.
Chi-Square Test-The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables.


The number of accidents per week in a city are as follows: 12,8,20,2,14,10,15,6,9,4. Are these frequencies in agreement with the belief that accident conditions were the same during this 10 weeks period?Use 0.05 sigificance level.
H0: Accident conditions are same during 10 weeks period 
H1: Accident conditions are not same during 10 weeks period
alpha=0.05 
Test statistic, Chi-square


```python
from numpy import sqrt, abs, round
# Given Values
Observed_Values=[12,8,20,2,14,10,15,6,9,4]
print(Observed_values)

Expected_values=[10,10,10,10,10,10,10,10,10,10]
print(Expected_values)

dof=len(Observed_Values)-1
alpha = 0.05
# via tabulated chi square values
def chi_square_calc(Observed_values,Expected_values,dof,alpha):
    chi_square_cal=sum([(o-e)**2./e for o,e in zip(Observed_values,Expected_values)])
    chi_square_tab=scipy.stats.chi2.ppf(1-alpha, df=dof)
    return round(chi_square_cal,3),round(chi_square_tab,4)
chi_square_cal,chi_square_tab=chi_square_calc(Observed_values,Expected_values,dof,alpha)
print(chi_square_cal,chi_square_tab)
if(chi_square_cal<chi_square_tab):
    print("Accept Null Hypothesis")
else:
    print("Reject NUll Hypothesis")
```
